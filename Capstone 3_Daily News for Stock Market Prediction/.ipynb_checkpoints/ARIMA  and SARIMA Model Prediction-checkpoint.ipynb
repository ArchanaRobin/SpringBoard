### Making Time series stationary

 Run the augmented Dicky-Fuller test on the time series to test for stationarity.

```python
results = price["Adj Close"]
```

```python
# Import the adfuller module from statsmodels.
from statsmodels.tsa.stattools import adfuller
# Run the Augmented Dickey-Fuller test on the series of closing stock prices, which is the column 'Adj Close'.
results1 = adfuller(results, autolag='AIC')

# Print out just the p-value of the test (results[0] is the test statistic, and results[1] is the p-value).
print("ADF test statistic : "  + str(results1[0]))
print("p-value : "  + str(results1[1]))
for key, value in results1[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}') 
```

```python
from statsmodels.tsa.stattools import kpss
kpss(results)
```

#### Observation:
    The p-value is obtained is greater than significance level of 0.05 and the ADF statistic is higher than any of the critical values. Clearly, there is no reason to reject the null hypothesis. So, the time series is in fact non-stationary.

### Seasonality Differencing
Many time series exhibit strong seasonal behavior. For seasonal adjustments, instead of taking first differences, you will take differences with a lag corresponding to the periodicity.

```python
price['Price First Difference'] = results - results.shift(1)
```

```python
results.shift(1)
```

```python
price['Seasonal First Difference'] = results - results.shift(12)
```

```python
price.head(12)
```

```python
## Again test dickey fuller test
# Run the Augmented Dickey-Fuller test on the series of closing stock prices, which is the column 'Adj Close'.
results_diff = adfuller(price['Seasonal First Difference'].dropna())

# Print out just the p-value of the test (results[0] is the test statistic, and results[1] is the p-value).
print("ADF test statistic : "  + str(results_diff[0]))
print("p-value : "  + str(results_diff[1]))
for key, value in results_diff[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}') 
```

#### observation:
  - strong evidence against the null hypothesis, reject the null hypothesis. Data has no unit root and is stationary
 - The p-value is very less than the significance level of 0.05 and hence we can reject the null hypothesis and take that the series is stationary.
 - The p-value is extremely small, so we can easily reject the hypothesis that prices are a random walk at all levels of significance.
 - Letâ€™s visualise the series as well to confirm.

### Autocorrelation and Partial Autocorrelation

 - Identification of an AR model is often best done with the PACF. 

    - The number of non-zero partial autocorrelations gives the order of the AR model. By the `order of the model` we mean the most extreme lag of x that is used as a predictor.
    

 - Identification of an MA model is often best done with the ACF rather than the PACF.
 

   - For an MA model, the theoretical PACF does not shut off, but instead tapers toward 0 in some manner. A clearer pattern for an MA model is in the ACF. The ACF will have non-zero autocorrelations only at lags involved in the model.

    - p,d,q p AR model lags d differencing q MA lags

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(price['Seasonal First Difference'].iloc[13:],lags=40,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(price['Seasonal First Difference'].iloc[13:],lags=40,ax=ax2)
```

 -  p is the order of the AR term

 -  is the order of the MA term

 - d is the number of differencing required to make the time series stationary

 - If the time series is already stationary, then d = 0.

#### Second method to make data Stationary - Log Transformation:

A classic way of transforming stock time series is the log-return of the series. This is calculated as follows:
$$ \log\text{_return} (y_t) = \log \big(\frac{y_t}{y_{t-1}}\big)$$

```python
results_log = np.log(results)
```

```python
results_log.plot(grid=True)
```

```python
# Create a new DataFrame of seasonally adjusted earnings by taking the lag-4 difference of quarterly earnings using the .diff() method.
results_log = results_log.diff()

# Examine the first 10 rows of the seasonally adjusted DataFrame and notice that the first four rows are NaN.
results_log.head(10)

# Drop the NaN rows using the .dropna() method.
results_log = results_log.dropna()
results3 = kpss(results_log)
print("p value : ",  results3[1])
print("test statistic : "  + str(results3[0]))

#Plot the transformes time series
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(21,6))
ax1.plot(results_log)
plot_acf(results_log, ax = ax2)
plot_pacf(results_log, ax = ax3)
plt.show()
```

#### Observation:
     Our p-value 0.1 is now greater than 0.05, so we can accept the null hypothesis that our data is stationary.

Note: Now the time series is ready for miodeling.

## Modeling - ARIMA
- ARMA models
    - Autoregressive moving-average (ARMA) model
    - ARMA = AR + MA
        - ARMA(1, 1) model:
$$ y_t = a_1 y_{t-1} + m_1 \epsilon_{t-1} + \epsilon_t $$
        - ARMA(p, q) model:
            - p is order of AR part
            - q is order of MA part

 - MA models
    - Moving Average (MA) model
        - MA(1) model:
$$ y_t = m_1 \epsilon_{t-1} + \epsilon_t $$
        - MA(2) model:
$$ y_t = m_1 \epsilon_{t-1} + m_2 \epsilon_{t-2} + \epsilon_t $$
        - MA(q) model:
$$ y_t = m_1 \epsilon_{t-1} + m_2 \epsilon_{t-2} + \cdots + m_q \epsilon_{t-q} + \epsilon_t $$ 

#### Fitting an ARMA model

 - This sets the lag value to 3 for autoregression, uses a difference order of 1 to make the time series stationary, and uses a moving average model of 3.
 
   - p is the order of the AR term

   - q is the order of the MA term
   
   - d is the number of differencing required to make the time series stationary
   
   If the time series is already stationary, then d = 0.

```python
model3 = ARIMA(X_train, order=(0, 1, 3))
results3 = model3.fit()
MSE = np.mean(np.abs(results3.resid))
# Print the MAE.
print(MSE)
# Use the DataFrame's .plot() method with no arguments to plot the  time series.
price.plot();
plt.show()
```

```python
history = [x for x in X_train]
predictions = []
observations = len(X_test)

for time_point in range(observations):
    model = ARIMA(history, order=(0,1,3))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    true_test_value = X_test[time_point]
    history.append(true_test_value)

MSE_error = mean_squared_error(X_test, predictions)
print('Root Mean Squared Error is {}'.format(np.sqrt(MSE_error)))
```

```python
# Plot the predictions against the test data
test_range = X_test.index
plt.plot(test_range, predictions, color='blue', marker='o', linestyle='dashed',label='Predicted Price')
plt.plot(test_range, X_test, color='red', label='Actual Price')
```

```python
# Plot the full train and test data sets and predictions
test_set_range = X_test.index
y_set_range = X_train.index
plt.plot(X_train, color='blue', marker='o', linestyle='dashed',label='Predicted Price')
plt.plot(test_set_range, X_test, color='red', label='Actual Price')
plt.plot(test_set_range, predictions, color='green', label='Predicted Price')
```

#### Unrolling ARIMA forecast
ARIMA in order to forecast the absolute value of the dataset. Sometimes predicting the difference could be enough; will the stocks go up, or down; but sometimes the absolute value is key.

```python
from numpy import cumsum
# Use the .get_forecast() method of the model_fit object and select the predicted mean of the next 10 differences.
arima_diff_forecast = model_fit.get_forecast(step=10).predicted_mean

# Use the np.cumsum() function to integrate your difference forecast.
arima_int_forecast = np.cumsum(arima_diff_forecast) 

# Add the last value of the original DataFrame to make your forecast an absolute value.
arima_value_forecast = arima_int_forecast + price.iloc[-1, 0]

print(arima_value_forecast)
```
